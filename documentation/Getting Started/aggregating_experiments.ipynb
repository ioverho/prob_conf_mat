# Aggregating Experiments

```python
confusion_matrix_a = [
    [9, 1, 0, 3, 0, 0, 0],
    [2, 52, 2, 3, 0, 1, 0],
    [0, 4, 17, 6, 0, 0, 0],
    [1, 6, 1, 138, 0, 0, 0],
    [0, 2, 0, 2, 20, 0, 1],
    [0, 5, 0, 0, 0, 9, 1],
    [0, 2, 2, 4, 1, 0, 27],
]

confusion_matrix_b = [
    [6, 1, 0, 2, 1, 1, 0],
    [3, 60, 2, 6, 1, 1, 1],
    [0, 4, 9, 6, 1, 0, 1],
    [1, 8, 1, 118, 0, 3, 2],
    [0, 1, 1, 1, 7, 1, 2],
    [1, 4, 2, 1, 3, 10, 2],
    [2, 4, 3, 5, 4, 0, 27],
]

confusion_matrix_c = [
    [7, 2, 0, 2, 1, 1, 1],
    [5, 41, 4, 6, 1, 3, 2],
    [1, 3, 10, 5, 2, 1, 0],
    [1, 5, 2, 142, 4, 0, 1],
    [4, 0, 1, 3, 9, 1, 1],
    [1, 2, 1, 2, 0, 6, 1],
    [1, 1, 2, 4, 1, 2, 23],
]

confusion_matrix_d = [
    [7, 2, 0, 3, 0, 0, 0],
    [1, 53, 1, 3, 0, 0, 0],
    [0, 5, 18, 6, 0, 0, 0],
    [0, 5, 0, 139, 0, 0, 0],
    [0, 3, 0, 3, 21, 0, 1],
    [0, 6, 0, 0, 0, 10, 0],
    [0, 0, 5, 2, 2, 0, 25],
]
```

```python
from bayes_conf_mat import Study

study = Study(
    name="comparing_experiments",
    seed=0,
    num_samples=10000,
    ci_probability=0.95,
)

study.add_experiment(
    "a/test_a",
    confusion_matrix_a,
)

study.add_experiment(
    "b/test_b",
    confusion_matrix_b,
)

study.add_experiment(
    "c/test_c",
    confusion_matrix_c,
)

study.add_experiment(
    "d/test_d",
    confusion_matrix_d,
)

study.add_metric(metric="ba")
study.add_metric(metric="ba+adjusted=True")
study.add_metric(metric="mcc")
study.add_metric(metric="dor")

study.plot_metric_summaries(
    metric="ba+adjusted=True",
    class_label=2,
    normalize=True,
    base_line_width=1,
    extrema_lines_width=1,
);
```
